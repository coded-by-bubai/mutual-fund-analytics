# Exploratory Data Analysis (EDA) Findings


## Key Insights


- Daily NAV showed a consistent upward trend across most equity mutual funds between 2022 and 2025, indicating long-term market growth despite short-term volatility.

![daily nav trend](../reports/nav_trend_matplotlib.png)

- Assets Under Management (AUM) increased steadily from 2022 to 2025, with SBI Mutual Fund maintaining the highest AUM among all fund houses.
![AUM growth](../reports/aum_growth_by_amc.png)

- Monthly SIP inflows demonstrated continuous growth throughout the analysis period, reaching their highest level in December 2025, indicating increasing investor participation.
![SIPinflow](../reports/sip_inflow_timeseries.png)

- Mutual fund folio counts nearly doubled between January 2022 and December 2025, reflecting expanding retail investor adoption.
![](../reports/folio_count_growth.png)

- Investors in the 26–35 and 36–50 age groups contributed the largest share of SIP investments, while higher SIP amounts were generally observed among older investors.
![](../reports/investor_age_distribution.png)

- Metropolitan (T30) cities contributed a larger share of total SIP investments compared to B30 cities, although B30 participation showed gradual improvement over time.
![](../reports/t30_b20_sip_distribution.png)

- Madhya Pradesh recorded the highest SIP contribution among the listed states, while Maharashtra had the lowest, indicating significant regional variation in mutual fund participation.
![](../reports/sip_by_state.png)

- Correlation analysis revealed that large-cap equity funds had strong positive correlations in daily returns, indicating similar market behaviour and diversification limitations within the same category.
![](../reports/correlation_matrix.png)

- Financial Services, Information Technology, and Oil & Gas were the largest sector allocations across equity fund portfolios, highlighting the concentration of investments in major sectors of the Indian economy.
![](../reports/sector_allocation_donut.png)

- Liquid Funds consistently attracted the highest net inflows throughout the year (32,374–41,952 Cr), significantly outperforming all other categories, highlighting strong investor preference for liquidity and short-term investments.
![](../reports/category_heatmap.png)

## Overall Conclusion

The mutual fund industry experienced significant growth during 2022–2025, driven by increasing SIP inflows, rising AUM, expanding investor participation, and sustained growth in equity-oriented investments. The analysis indicates strong market confidence, continued retail participation, and consistent long-term performance of diversified equity mutual funds.
